```python
# y = sin(x)
import math
x = np.linspace(0,2*math.pi,100)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.grid()
```

```python
x = np.arange(0,4*np.pi,0.1)
plt.plot(x,np.sin(x))
plt.axhline(y=0)
plt.axvline(x=2*np.pi)
```

```python
time        = np.arange(0, 10, 0.1);

amplitude   = np.sin(time)

plt.plot(time, amplitude)

plt.title('Sine wave')

plt.xlabel('Time')

plt.ylabel('Amplitude = sin(time)')

plt.axhline(y=0, color='k')

plt.show()
```
